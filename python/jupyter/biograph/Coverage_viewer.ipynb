```python
"""
    View reference coverage for multiple individuals for a given region.
"""
from biograph import BioGraph, Reference, find_variants
```

```python
# Import seqset and readmap
my_bg = BioGraph("/share/datasets/HG002/HG002-NA24385-50x.bg/")
```

```python
# Import a reference
grch37 = Reference("/share/reference/human_g1k_v37/")
```

```python
# Make pick an area of interest
chromosome = "6"
start = 10250000
end = 10255000
```

```python
# Find variants and coverage for this area
v = find_variants(my_bg, grch37, chromosome, start, end)[0]
```

```python
%matplotlib inline
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(14,6))

coverage, = plt.plot(v.coverage, color='lightblue', linewidth=2)

plt.xlabel('Position')
plt.ylabel('Coverage')
plt.title('Depth of reference and alt sequence coverage by position, {chr}:{start}-{end}'.format(
        chr=v.ref_range.scaffold, start=v.ref_range.start, end=v.ref_range.end
    )
)
plt.grid(True)

plt.legend([coverage], ['Reference coverage'])

size = len(v.coverage)
ax.set_xticks(np.arange(size, step=int(size * 0.1)))
ax.set_xticklabels(np.arange(v.ref_range.start, v.ref_range.end, step=int(size * 0.1)))

ax.margins(0, 0.1)
```
